```python
import re
import PyPDF2

pdf1 = open('Pengumuman_Harta_Kekayaan_LHKPN_27412 (1).pdf', 'rb')
reader1 = PyPDF2.PdfReader(pdf1)

num_pages = len(reader1.pages)

# Loop through the pages of the PDF file and extract the text content
text = ''
for i in range(num_pages):
    page = reader1.pages[i]
    text += page.extract_text()

# Define the regex patterns to search for key-value pairs
patterns = {
    'Nama': r'Nama\s*:(.*)',
    'Jabatan': r'Jabatan\s*:(.*)',
    'NHK': r'NHK\s*:(.*)',
    'Bidang':r'BIDANG\s*:(.*)',
    'Lembaga':r'LEMBAGA\s*:(.*)',
    'Unit Kerja':r'UNIT KERJA\s*:(.*)'
}

data = {}

# Search for key-value pairs using regex patterns
for key, pattern in patterns.items():
    match = re.search(pattern, text)
    if match:
        value = match.group(1).strip()
        data[key] = value

# Print the extracted data
for key, value in data.items():
    print(f'{key}: {value}')
```

```python
import os
import PyPDF2
import json

folder_path = "sample"  # Path folder dengan file-file PDF

data_list = []

# Loop melalui setiap file dalam folder
for filename in os.listdir(folder_path):
    if filename.endswith(".pdf"):
        file_path = os.path.join(folder_path, filename)

        with open(file_path, "rb") as pdf_file:
            reader = PyPDF2.PdfReader(pdf_file)
            num_pages = len(reader.pages)

            # Loop through the pages of the PDF file and extract the text content
            text_combined = ""
            for i in range(num_pages):
                page = reader.pages[i]
                text = page.extract_text()
                text_combined += text

            data = {}
            nama = text_combined.split("\n")[7].split(":")[1]
            lembaga = text_combined.split("\n")[3].split(":")[1]
            unit_kerja = text_combined.split("\n")[4].split(":")[1]
            jabatan = text_combined.split("\n")[8].split(":")[1]
            nhk = text_combined.split("\n")[9].split(":")[1]
            tanah = int(text_combined.split("\n")[11].split(":")[0].split("Rp. ")[1].replace(".", ""))

            data["Nama"] = nama
            data["Lembaga"] = lembaga
            data["Unit Kerja"] = unit_kerja
            data["Jabatan"] = jabatan
            data["NHK"] = nhk
            data["Tanah dan Bangunan"] = tanah

            data_list.append(data)

json_data = json.dumps(data_list, indent=2)
print(json_data)
```

```python
# mengexport semua halaman pdf menjadi gambar
import fitz

def ambil_gambar_dari_pdf(nama_file_pdf):
    doc = fitz.open(nama_file_pdf)
    jumlah_halaman = doc.page_count

    for halaman in range(jumlah_halaman):
        objek_halaman = doc.load_page(halaman)
        objek_gambar = objek_halaman.get_pixmap()

        if objek_gambar is not None:
            data_gambar = objek_gambar.tobytes()

            # Simpan data gambar ke file
            nama_file_gambar = f"gambar_halaman_{halaman+1}.jpg"
            with open(nama_file_gambar, "wb") as file_gambar:
                file_gambar.write(data_gambar)

    doc.close()

# Contoh penggunaan
file_pdf = "Housing_Statistics_December_2021.pdf"

ambil_gambar_dari_pdf(file_pdf)
```

```python
#mengambil gambar pada pdf (semua gambar)

import fitz

def extract_images_from_pdf(pdf_path, output_dir):
    doc = fitz.open(pdf_path)
    image_count = 0

    for page_num, page in enumerate(doc):
        image_list = page.get_images()

        for image_index, img in enumerate(image_list):
            xref = img[0]
            base_image = doc.extract_image(xref)
            image_data = base_image["image"]

            image_format = base_image["ext"]
            image_filename = f"image_{page_num + 1}_{image_index + 1}.{image_format}"
            image_path = os.path.join(output_dir, image_filename)

            with open(image_path, "wb") as image_file:
                image_file.write(image_data)

            image_count += 1

    doc.close()
    return image_count


# Contoh penggunaan
pdf_file = "lengkap ENSIKLOPEDIA.pdf"
output_directory = "output"

total_images = extract_images_from_pdf(pdf_file, output_directory)
print(f"Jumlah gambar yang berhasil diambil: {total_images}")
```

```python
import os
from PyPDF2 import PdfReader
from pdf2image import convert_from_path

def extract_images_by_keywords(pdf_path, output_dir, keywords):
    # Membaca file PDF
    pdf = PdfReader(pdf_path)

    # Membuat direktori output jika belum ada
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Mencari halaman dengan kata kunci dan mengekstrak gambar
    for page_num, page in enumerate(pdf.pages):
        content = page.extract_text().lower()

        # Memeriksa keberadaan kata kunci dalam teks halaman
        if any(keyword in content for keyword in keywords):
            # Mengonversi halaman PDF menjadi gambar
            images = convert_from_path(pdf_path, first_page=page_num+1, last_page=page_num+1, output_folder=output_dir)

            # Menyimpan gambar
            for i, image in enumerate(images):
                image_path = os.path.join(output_dir, f'image_{page_num}_{i}.png')
                image.save(image_path, 'PNG')

# Contoh penggunaan
pdf_file = 'Housing_Statistics_December_2021.pdf'
output_directory = 'output'
search_keywords = ['grafik', 'Chart']  # Kata kunci untuk mencari halaman dengan gambar
extract_images_by_keywords(pdf_file, output_directory, search_keywords)
```

```python
from PyPDF2 import PdfReader
import json

def extract_text_from_pdf(pdf_file):
    pdf = PdfReader(pdf_file)

    extracted_text = []

    for page in pdf.pages:
        text = page.extract_text()
        extracted_text.append(text)

    return extracted_text

def save_text_as_json(text, json_file):
    with open(json_file, 'w') as file:
        json.dump(text, file)

# Contoh penggunaan
pdf_file = "Pengumuman_Harta_Kekayaan_LHKPN_27412 (1).pdf"
json_file = "output3.json"

extracted_text = extract_text_from_pdf(pdf_file)
save_text_as_json(extracted_text, json_file)
```
