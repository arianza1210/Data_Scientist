**HARGA RUMAH JAKSEL**
Dataset harga rumah jaksel terdiri dari 7 kolom dengan jumlah data yaitu 1000 data. Kolom tersebut terdiri dari :


1.   HARGA : harga dari rumah.
2.   LT : jumlah luas tanah.

1.   LB : jumlah luas bangunan.
2.   JKT : jumlah kamar tidur.

1.   JKM : jumlah kamar mandi.
2.   GRS : ada/tidak ada

1.   KOTA : nama kota.

Preprocessing

```python
df.drop(columns='KOTA', inplace=True, axis=1)
```

```python
df.GRS=df.GRS.map({'ADA':1,'TIDAK ADA':0})
```

```python
df.head()
```

```python
#MEchine Learning Modeling
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

```python
X=df.iloc[:,1:]
X
```

```python
y=df.iloc[:,0]
y
```

```python
X=df.drop('HARGA',axis=1)
y=df['HARGA']
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
import numpy as np

model= RandomForestRegressor()
model.fit(X_train,y_train)
hasil=model.predict(X_test)
```

```python
hasil
```

```python

print("Akurasi Training : ",model.score(X_train, y_train))
print("Akurasi Testing : ", model.score(X_test,y_test))
```

```python
def prediksi(LT,LB,JKT,JKM,GRS):
    predict = pd.DataFrame()
    predict['LT'] = [LT]
    predict['LB'] = [LB]
    predict['JKT'] = [JKT]
    predict['JKM'] = [JKM]
    predict['GRS'] = [GRS]
    print(model.predict(predict))
```

```python
prediksi(754,500,4,4,1)
```
